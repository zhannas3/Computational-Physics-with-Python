# Assignment 6. Interpolation

This assignment investigates Lagrange and cubic spline interpolations for the given function: $f(x)=\frac{1}{(1+25x^2)}$. 
The function takes values for x between -1 and 1. The lagrange polynomials and cubic splines were generated at points x=-1,-0.75,-0.5,-0.25,0,0.25,0.5,0.75,1. 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
def f(x):
    return 1/(1+25*x*x)
```

Below is the given points at which polynomials and cubic splines should be plotted.

```python
x=np.arange(-1,1.25,0.25)
plt.plot(x,f(x),'o',label="$f(x)=1/(1+25x^2)$")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.grid()
plt.show()
```

Below is the implementation of lagrange polynomial:

```python
def lagrange(x,y,n):
    s=0
    for k in range(len(y)):
        p1=1
        p2=1
        for i in range(len(x)):
            if i==k:
                p1=p1*1
                p2=p2*1
            else:
                p1=p1*(n-x[i])
                p2=p2*(x[k]-x[i])
        s=s+y[k]*p1/p2
    return s
```

We set up the points to be 100 equally spaced points in the interval [-1,1] and then plot the interpolation polynomials. Here we have a polynomial of 8th order:

```python
x=np.array([-1,-0.75,-0.5,-0.25,0,0.25,0.5,0.75,1])
y=f(x)
xnew=np.linspace(-1,1,100)
ynew=[lagrange(x,y,i) for i in xnew]
plt.figure(figsize=(9,5))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,'o',xnew,ynew, label="8th order")
plt.grid()
plt.legend()
plt.show()
```

```python
error=max(abs(ynew-f(xnew)))
print("Error is:",error)
```

```python
def LUdecomp(c,d,e):
    n = len(d)
    for k in range(1,n):
        factor = c[k-1]/d[k-1]
        d[k] = d[k] - factor*e[k-1]
        c[k-1] = factor
    return c,d,e
def LUsolve(c,d,e,b):
    n = len(d)
    for k in range(1,n):
        b[k] = b[k] - c[k-1]*b[k-1]
    b[n-1] = b[n-1]/d[n-1]
    for k in range(n-2,-1,-1):
        b[k] = (b[k] - e[k]*b[k+1])/d[k]
    return b
```

```python
def curvatures(x,y):  #returns curvatures at nodes
    n = len(x) - 1
    c = np.zeros(n)
    d = np.ones(n+1)
    e = np.zeros(n)
    k = np.zeros(n+1)
    c[0:n-1] = x[0:n-1] - x[1:n]
    d[1:n] = 2.0*(x[0:n-1] - x[2:n+1])
    e[1:n] = x[1:n] - x[2:n+1]
    k[1:n] =6.0*(y[0:n-1] - y[1:n]) \
            /(x[0:n-1] - x[1:n]) \
            -6.0*(y[1:n] - y[2:n+1]) \
            /(x[1:n] - x[2:n+1])
    LUdecomp(c,d,e)
    LUsolve(c,d,e,k)
    return k

def Spline(x,y,k,x0):  #gives interpolant at value x0
    def findSegment(x,x0):   #finds the portion of the spline that contains x0
        Left = 0
        Right = len(x)- 1
        while 1:
            if (Right-Left) <= 1: 
                return Left
            i =(Left + Right)/2
            if x0 < x[i]: 
                 Right = i
            else: 
                Left = i
        i = findSegment(x,x0)
        h = x[i]-x[i+1]
        y = ((x0 - x[i+1])**3/h -(x0-x[i+1])*h)*k[i]/6.0 \
        - ((x0 - x[i])**3/h-(x0-x[i])*h)*k[i+1]/6.0 \
        + (y[i]*(x0 - x[i+1]) \
        - y[i+1]*(x0 - x[i]))/h
    return y
```
