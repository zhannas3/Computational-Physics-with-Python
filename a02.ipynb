# Assignment 2. Numerical Errors

The assignment's main objective is to estimate truncation errors which are happened by truncating an infinite sum and approximating by finite sum. The experiment was done on investigating the exponential expansion using Maclaurin series, which is given by the following formula:
                              $$e^x=\sum_{n=0}^{\infty}\frac{x^n}{n!}$$

Firstly, all needed libraries were imported: "random" and "matplotlib.pyplot".

```python
from math import factorial
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

#below the true values of e^x in 16 digits
y=2.7182818284590452**1
z=2.7182818284590452**2
k=2.7182818284590452**10
```

### Part a) and c) combined. (x=1)

The following function calculates the values of $e^1$ when terms were added sequentially. It contains two variables, x - the point at which we need to estimate the value of $e^1$, pres - the needed precision.

```python
def macExp(x,pres):
    n=[]          #number of terms
    error=[]      
    ex=[]         #set of e^x values when terms were added sequentially
    exp1=0.0
    for i in range(0,100):       #range was chosen from 0 to 100
        exp1=exp1+(x**i)/factorial(i)
        if (exp1 != y):
            ea=abs((y-exp1)/y)     
            n.append(i)
            ex.append(exp1)
            error.append(ea)
            temp=y-exp1  
            #temporary value that stores the difference of true and actual values
        if temp<pres:            
            break         
    return n, ex, error
```

Part c) The function also counted the needed number of terms to reach a precision $10^{-15}$. It was shown that it is sufficient to have **18 terms**. 

### Part b) The relationship between error vs. number of terms 

True relative error was calculated inside the for loop:
                                 $$error=\left|\frac{x-x'}{x'}\right|\times100\%$$
- where, $x$ means the actual value;
- $x'$ means true value.

The analysis was done based on all given data of number of terms and error values. The graph of error as a function of the number of terms was plotted. The graph suggests that error is decreasing starting from 63.2 (in usual scale) after each iteration.

```python
n1,ex1,error1=macExp(1,1e-15)
plt.plot(n1,error1,label="Error vs. Number of terms")
plt.legend()
plt.xlabel("$Number \quad{of} \quad{terms}$",size=15) 
plt.ylabel("$Error$", size=15)
plt.yscale('log')
plt.grid()
plt.show()
```
