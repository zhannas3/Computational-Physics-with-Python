In order to find an average: $$avg=S/N$$ was used. Where, S - sum of all elements, N - number of elements.

```python
#first found the sum of the elements in lists
def total(x):
    s=0
    for i in range(500):
        s=s+x[i]
    return s
```

```python
%%time
for i in range(500):
    time.sleep(0.001)
def average(z):
    avg=float(total(z)/500)    #since there are 500 elements
    return avg
```
